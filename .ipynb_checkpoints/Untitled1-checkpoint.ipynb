# Libraries imported

```python
# Librarires are imported
from textblob import TextBlob 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from collections import Counter
sns.set()
```

Load the twentieth-century data that you scraped in Exercise 1.4.

```python
# Import txt file

with open('20th_Century_Wewbpage.txt', 'r', errors='ignore') as file:
    data = file.read().replace('\n', '')
```

# Tokenization

Tokenize the words from the text and create a bar chart to plot the 10 most common words.

```python
import nltk
nltk.download('punkt_tab')
```

```python
# Sentence tokenization 

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent)
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# Create frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words)
```

```python
dist_words.most_common(10)
```

```python
# Plot the 30 most common words
plt.figure(figsize=(12, 6))
nltk.FreqDist(tokenized_word).plot(30, cumulative=False)
plt.show()
```

# Stop Words

Remove stop words (and punctuation marks if necessary), and rerun the plot. What has changed? Comment on your impressions in a markdown cell under the plot.

```python
# Defining stopwords

from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
print(stop_words)
```

```python
# Removing stopwords in sentences

# filtered_sent = []
# for word in tokenized_sent:
#     if word not in stop_words:
#         filtered_sent.append(word)
# print("Tokenized Sentence:", tokenized_sent)
# print("Filterd Sentence:", filtered_sent)
```

```python
# Removing stopwords in words

filtered_words = [] # creates an empty list
for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
filtered_words
```

```python
# Create a new FreqDist for filtered_words

dist_words_filter = FreqDist(filtered_words)
print(dist_words_filter)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative = False)
plt.show()
```

```python
dist_words_filter
```

```python
# Substitute all punctuations marks with a space 

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                          " ",          # Replace all non-letters with spaces
                          str(filtered_words))
```

```python
sans_punc
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print(tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(30, cumulative = False)
plt.show()
```

```python
dist_words_filter_2.most_common(10)
```

# Initially, the most common words were ('.', 1311), ('the', 1132), and (',', 975), which are not meaningful for analysis.
# After removing stop words and punctuation marks, the new most common words are ('The', 322), ('December', 223), and ('Retrieved', 207), which offer more meaningful insights for analysis. However, "The" still appears as the most frequent word, indicating that further cleaning is needed, as it does not contribute useful information.

# POS Tags with TextBlob

* Create a tags list using TextBlob
* Create a list of the Top 10 POS tags for words that appear in the article.
* Plot this list in a bar chart using the seaborn library.

```python
new_stopwords = ["And", "Then", 'n', 't', 's', 'The']
```

```python
filtered = []
for word in tokenized_word_2:
    if word not in new_stopwords:
        filtered.append(word)
```

```python
%%time
text = TextBlob(str(filtered))
```

```python
text
```

```python
import nltk
nltk.download('averaged_perceptron_tagger_eng')
```

```python
tags_list = text.tags
```

```python
tags_list
```

```python
df_text = pd.DataFrame(tags_list)
df_text.columns = ['Words', "Word type"]
```

```python
df_text.head()
```

```python
df_t = df_text.groupby('Word type').count().reset_index()
```

```python
df_t.head()
```

```python
top10 = df_t.nlargest(10, 'Words')
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 20):
    sns.barplot(x = "Words", y = "Word type",
    saturation = 0.9, data = top10).set_title("20th Century - top 10 word types used")
```

Create three bar plots with the top 15 POS labels—one each for nouns, verbs, and adjectives. Analyzing the results—what do you think stands out in these frequencies? Comment on your impressions in a Markdown cell under the plot.

# Nouns

```python
df = df_text[(df_text['Word type'] == "NN") | (df_text['Word type'] == "NNS") | (df_text['Word type'] == "NNP")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize=(10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data = top15).set_title("20th Century - Top 15 frequently used nouns")
```

By analyzing this set of words, it is possible to deduce that one of the most prominent topics in the 20th century was the World Wars, particularly World War II.

# Verbs

```python
df = df_text[(df_text['Word type'] == "VB")  | (df_text['Word type'] == "VBD")]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by = ['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize = (10, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x = "Word", y = "Occurences",
    saturation = 0.9, data = top10).set_title("20th Century - Top 15 frequently used verbs")
```

By examining the full text on the webpage, it becomes clear why the word "Retrieved" is among the most common, it appears frequently in the References section.

# Adjectives

```python
df = df_text[df_text['Word type'] == "JJ"]
df.columns = ["Word", "Occurences"]
x = df.groupby('Word').count().reset_index()
y = x.sort_values(by=['Occurences'], ascending=False)
top15 = y.nlargest(15, 'Occurences')
```

```python
top15
```

```python
plt.figure(figsize=(13, 5))
with sns.dark_palette("xkcd:blue", 10):
    sns.barplot(x="Word", y="Occurences",
    saturation=0.9, data=top15).set_title("20th Century - Top 15 frequently used verbs Adjectives")
```

The most common adjectives are closely related to the main topic of the webpage, which is World War II.
