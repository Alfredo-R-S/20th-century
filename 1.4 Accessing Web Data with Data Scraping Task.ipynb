# Scraping Wikipedia with requests and bs4

```python
# Second portion

import requests
from bs4 import BeautifulSoup
```

```python
# Get URL

page = requests.get("https://en.wikipedia.org/wiki/Key_events_of_the_20th_century")
```

```python
# Create soup and get title

soup = BeautifulSoup(page.text, 'html.parser')
print(soup.title)
```

```python
print(soup.text)
```

```python
text = soup.get_text()
```

```python
text = text.encode ('utf-8')
```

```python
# Save the page information as a text document
with open('20th_Century_Wewbpage.txt', 'wb') as f:
       f.write(text)
```
