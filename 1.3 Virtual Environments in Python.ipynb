```python
# Import Libraries
import pandas as pd
```

```python
# Random Data
data = pd.DataFrame({"Country": ['Mexico', 'Japan', 'Italy','USA'],
                     "Food": ["Tacos", "Sushi", "Pizza", "Hamburger"],
                     "Price": [3,12,10,11]})
```

```python
data
```

```python
# mathematical procedure using a lambda function
multiply = lambda x: x*10
multiply(data['Price'])
```

```python
# flag column
data['Price_Level'] = data['Price'].apply(
lambda x:  'high' if x > 10 else ('normal' if x >=5 else 'low'))
```
